```python
df_spy["delta^2"] = df_spy["delta"]
```

# Analysis
The plot above shows a comparative analysis of stock market sector performance between the 7/1/2019 and 6/30/2022 periods. As we can see in the above, the strongest performing sector over this period was Technology, represented by the S&P Sector SPDR Fund (XLK), while the worst performing sector for most of the period was Energy, represented by the S&P Sector SPDR Fund (XLE); for additional comparison, the entire S&P500 index is also shown as well as a generic bond index represented by the Vangaurd Total Bond Market Index Fund (BND). 

As we can see above, the Technology sector greatly outperformed the rest of the stock market leading up to and especially after the Covid Market Crash of March 2020, which can be explained by the effects of government lockdowns, business shutdowns, and the rise of remote work options for workers. Entities like Amazon became especially profitable during this period, as did the rise of "Uberfication" of services like food delivery.

Conversely, the Energy sector greatly underperformed the rest of the stock market for similar reasons up until about the beginning of 2022 - lockdowns, business shutdowns, and remote working opportunities greatly reduced demand for fuel in the transportation of goods/work commuting as well as airline travel, though the beginning of the war in Ukraine followed by the general post-Covid recovery greatly elevated Energy stocks due to the interruption in global supply chains along with the subsequent relative scarcity of oil. It should be especially noted that Energy stock nearly became the highest performing sector since the beginning of the 7/1/2019 period around the end of this dataset (6/30/2022), narrowly falling behind the still strongly performing Technology sector.

In our original hypothesis, we presumed that sectors like Healthcare would be strongly performing and Consumer Discretionary would be underperforming over the time period under consideration. Though the plot above shows that Consumer Discretionary is roughly the second best performing sector (for similar reasons as noted above for the Technology sector, though still trending very closely with the general S&P 500), we were surprised to see Healthcare not making any significantly different trend from the general S&P 500 due to Covid vaccine development as well as related treatments in the ongoing pandemic. This is more evident in the filtered chart below.
