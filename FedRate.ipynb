```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Study data files
rate_path = "FEDFUNDS.csv"
```

```python
rateDF = pd.read_csv(rate_path)
```

```python
rateDF.head()
```

```python
Date = rateDF['DATE']
funds = rateDF['FEDFUNDS']
plt.plot( Date , funds)
plt.title("Federal Fund Rate over Time")
plt.xlabel("Dates")
plt.ylabel("Rate")
plt.xticks(rotation=90)
plt.tick_params(axis='x', labelsize=6)
plt.show
```
