```python
# import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Download the CSV file from the dataset for GDP

csvpath = "Datasets/GDP_and_Components.csv"

GDP_df = pd.read_csv(csvpath, encoding="UTF-8")

GDP_df.head(10)
```

# Refine the data to a new dataframe with only the relevant columns

```python
refinedGDP_df = GDP_df[["Base Year", "Gross Domestic Product, Nominal, Undjusted, Domestic Currency"]]
refinedGDP_df.head()
```

# Perform quality test of dataframe

```python
type(refinedGDP_df)
```

```python
len(refinedGDP_df)
```

# Add percent change

```python
refinedGDP_df["pct_chg"] = refinedGDP_df["Gross Domestic Product, Nominal, Undjusted, Domestic Currency"].pct_change()
```

```python
refinedGDP_df['pct_chg'] = refinedGDP_df['pct_chg'] * 100
refinedGDP_df
```

# Plot the US GDP data

```python
plt.plot(refinedGDP_df["Base Year"],refinedGDP_df["Gross Domestic Product, Nominal, Undjusted, Domestic Currency"])
plt.title("US Gross Domestic Product Over Time", fontweight = "bold")
plt.xlabel("Date", fontweight = "bold")
plt.ylabel("GDP Currency USD, Trillions", fontweight = "bold")
plt.xticks(rotation=45)
plt.show()
```

GDP or gross domestic product can be defined as the monetary value of all finished goods and services produced within a country's borders and functions as a comprehensive scorecard of a country's economic health. 

I originally assumed the value of GDP would have remained relatively low during the time after its initial dip and before its healthy return back to pre-covid levels for at least a year. 

What I found was that this downtime of the US GDP being below the pre-covid level and was shorter than expected and also did not remain very low. 

It started to decrease around the 3rd quarter of 2019, and reached its minimum value in the second quarter of 2020. Its value fell by about 9% over this time which is quite dramatic.  But GDP value returned back to its pre-covid level around the 3rd quarter of 2020 following a rapid bounce back and continued to increase substantially. 

Some of these changes in dips and increases in value can obviously be attributed to businesses and factories closing and also government relief actions. 


```python
plt.plot(refinedGDP_df["Base Year"],refinedGDP_df["pct_chg"])
plt.title("Percent Change", fontweight = "bold")
plt.xlabel("Date", fontweight = "bold")
plt.ylabel("Percent Change", fontweight = "bold")
plt.xticks(rotation=45)
plt.show()
```
