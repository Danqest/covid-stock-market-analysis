```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
# Study data files
rate_path = "FEDFUNDS.csv"

```

```python
rateDF = pd.read_csv(rate_path)

# x_vals = pd.date_range(start='1/1/2019', end='2/1/2023', periods=len(rateDF))
# rateDF = ['DATE'] = x_vals
# print(x_vals)

rateDF.columns = rateDF.columns.str.replace("DATE", "Date")
rateDF.columns = rateDF.columns.str.replace("FEDFUNDS", "Fed Rate")

```

```python
x_vals = pd.date_range(start='1/1/2019', end='2/1/2023', periods=len(rateDF))
rateDF['Date'] = x_vals
# print(x_vals)
```

```python
rateDF.head()
```

```python
fig,ax = plt.subplots()
fig.set_size_inches(12,10)
plt.xticks(rotation=45)
ax.plot(rateDF["Date"], rateDF["Fed Rate"], color="coral")
ax.plot(rateDF["Date"], rateDF["Monthly Change"], color="purple")
plt.legend(['Fed Rate', 'Monthly Change'])
plt.title("Federal Fund Rate over Time", fontweight="bold")
plt.ylabel("Rate (%)", fontweight="bold")
plt.xlabel("Dates", fontweight="bold")
plt.grid(axis='y', linestyle='-', linewidth=1)
ax.set_yticks([4.5, 4.0, 3.5, 3.0, 2.5, 2.0, 1.5, 1.0, 0.5, 0, -0.5, -1.0, -1.5])
plt.show()


```

Analysis:
What did the Federal Government do to help influence the economy after the Covid crash?

Looking at the dataset and comparing the dates before and after Covid. The Federal Government dropped the Fed Rate (The lending rate that is determined when a bank lends money to another bank) to 0.65% from 1.58%, a 12% drop on March 1st, 2021. Then they dropped it again to 0.05%, the lowest in the Fed Rate History the next month. They did this to open the doors to lending and to try and help stimulate the economy. They kept it this low for two months and then raised it to 0.08% where it remained through most of the 'Covid' period. Finally rising in March 1st 2022 to 0.20%. The results of this year long low Fed Rate certainly stimulated the economy, and it might have overstimulated it because at this time we are facing a Fed Rate of 4.57% as of Feb 1st, 2023.
